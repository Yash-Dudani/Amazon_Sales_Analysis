```python
#courier status
plt.figure(figsize=(20,9))
sns.countplot(data=df,x='Courier Status',hue='Status')
plt.show()
```

#majority of the orders are shipped and delivered to the buyer through courier

```python
df.info()
```

```python
df['Category']=df['Category'].astype(str)
c_d=df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d,bins=15,edgecolor="black",color="yellow")
plt.xticks(rotation=0)
plt.show()
```

#Most of the buyers are purchasing t-shirts

```python
check_B2B=df['B2B'].value_counts()
plt.pie(check_B2B,labels=check_B2B.index,autopct='%1.1f%%')
plt.show()
```

#99.2% of buyers are retailers and 0.8% are wholesalers

```python
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available Size')
plt.show()
```

```python
#top 10 states by order count:
top10_state=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(20,10))
ax=sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x='ship-state')
for bars in ax.containers:
     ax.bar_label(bars)
plt.xticks(rotation=0)
plt.xlabel('State')
plt.ylabel('Order Count')
plt.show()
```

#most of the orders are from maharaashtra
